Declaração e atribuição de variáveis.
Função de saída print

```python
nome = "Fábio"
print(nome)
```

Os nomes de variáveis em Python são **case sensitive**, diferencia maiúsculas de minúsculas.

```python
nome_usuario = "Fábio"
print(nome_usuario)
```

Verificaremos aqui se realmente os identificadores (nomes) de variáveis em Python são case sensisitve.

```python
nome_usuario = "Fábio"
Nome_usuario = "Gustavo"
print(nome_usuario)
print(Nome_usuario)
teste = (nome_usuario != Nome_usuario)
print("\nResultado do primeiro teste:", teste)
teste2 = (nome_usuario == Nome_usuario)
print("\nResultado do segundo teste:", teste2)
```

Declaração de variável do tipo numérica.

Nesse exemplo usamos também f-string.

```python
media = 0
print(media)
print(f'O valor da variável media é {media}')
```

Podemos declarar várias variáveis de mesmo valor ao mesmo tempo.

```python
n1 = n2 = n3 = n4 = 0.0

print('O valor de n1 é ', n1)
print('O valor de n2 é ', n2)
print('O valor de n3 é ', n3)
print('O valor de n4 é ', n4)
```

Podemos, ainda, criar mais de uma variável com tipos de dados diferentes e valores diferentes.

```python
nome, idade = 'Fábio', 47
print(nome)
print(idade)
```

Os tipos de dados são classificações dos dados que utilizamos nas variáveis dentro dos programas.

Criamos variáveis com 3 tipos de dados diferentes (int, float, str).

int: inteiro (variáveis media e idade)  
float: ponto-flutuante (variáveis n1, n2, n3 e n4)  
str: string, cadeia de caracteres (variável nome)

```python
media = 0
n1 = n2 = n3 = n4 = 0.0
nome, idade = 'Fábio', 47
```

Função type()

```python
media = 0
n1 = n2 = n3 = n4 = 0.0
nome, idade = 'Fábio', 47
estado = True # tipo de dado do tipo lógico ou booleano
complexo = 1+2j # tipo de dado do tipo número complexo ou complex

# Sintaxe para a função type()

type(media)
type(n1)
type(nome)
type(idade)
type(estado)
type(complexo)

# Imprimindo os retornos da função type()
# Os tipos em Python são chamados de classes

print('Tipo da variável media:')
print(type(media))

print('\nTipo da variável n1:')
print(type(n1))

print('\nTipo da variável nome:')
print(type(nome))

print('\nTipo da variável idade:')
print(type(idade))

print('\nTipo da variável estado:')
print(type(estado))

print('\nTipo da variável complexo:')
print(type(complexo))
```

Função isinstance() também é utilizada para nos informar o tipo de dado verificando se uma determinada afirmação sobre aquele dado é verdadeira.

```python
a = 10
b = 'Sol'

print(isinstance(a, int))
# vai pegar a variável "a" e verificar se ela é do tipo inteiro
```

A função isinstance() também verifica se uma variável é de um tipo dentro de um conjunto de tipos.

```python
a = 10

print(isinstance(a, (int, float)))
# verifica se a variável "a" é do tipo int ou do tipo float
```

Podemos também criar variáveis que recebem o conteúdo de cálculos ou concatenações de strings (cadeias de caracteres).

```python
a = 40
c = 3
r = a * c

print(r)
```
