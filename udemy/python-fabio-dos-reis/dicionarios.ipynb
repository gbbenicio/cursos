# Dicionários

Permite armazenar em pares chave, valor. São similares ao arrays associativos que existem em outras linguagens ou hashmaps.

Não permitem itens duplicados mas os itens podem ser de qualquer tipo.

A chave é uma string que dá acesso a um valor armazenado.  
As chaves são imutáveis.

```python
elemento = {
    'z': 3,
    'nome': 'Lítio',
    'grupo': 'Metais Alcalinos',
    'densidade': 0.534
}

print(f'Elemento: {elemento["nome"]}')
```
