# WHILE

Loops ou laços de repetição são usados para repetir blocos de códigos diversar vezes de forma automatizada, quantas vezes forem necessárias.

Em Python nós temos 2 blocos ou estruturas de repetição principais que são o WHILE e o FOR.

```python
num = 1

while (num <= 10):
  print(num)
  num += 1
print('Laço encerrado!')
```

O operador += 1 é um operador de atribuição com incremento.

Ele vai tomar o que está na variável **"num"** e acrescentar **1** ao que está lá dentro.

```python
Exercício:

Criar um programa que:
Vai perguntar o nome do usuário e escrever uma mensagem de boas-vindas.
Só encerra quando digitar a tecla "x".
```

```python
nome = None

while True:
  print('Digite seu nome, ou x para parar: ')
  nome = input()
  if nome == 'x' or nome == 'X':
    break # finaliza imediatamente o laço de repetição onde ela está inserida
  print(f'Bem-vindo, {nome}')
print('Até logo!')
```
