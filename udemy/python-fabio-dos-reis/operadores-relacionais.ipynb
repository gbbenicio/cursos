# Operadores Relacionais

Trataremos de operadores de comparação.

| Operador | Operação |
|----------|----------|
| == | Igual a |
| != | Diferente de |
| > | Maior que |
| < | Menor que |
| >= | Maior ou igual |
| <= | Menor ou igual |



```python
a = 5
b = 5
```

```python
a == 5
```

```python
b == 6
```

```python
a == b
```

```python
a != b
```

```python
b != a + 1
```

```python
a * 2 >= b
```

```python
print('Digite um número: ')
n1 = int(input())

n2 = int(input('Digite outro número: '))
```

\n é um caractere de escape e simboliza uma quebra de linha.

```python
x = n1 == n2
print('São iguais? ', x, '\n')
```

```python
z = n1 > n2
print(n1, 'é maior que', n2, '?', z, '\n')
```

```python
y = n1 != n2
print('São diferentes? ' + str(y))
```

O + neste caso é um operador de concatenação, que junta cadeia de caracteres (strings).

Só aceita operandos que sejam strings.

Como y é um booleano temos que converter para string e obtemos isso com o método str.
