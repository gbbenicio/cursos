# Funções  

Uma função é um bloco de código que executa uma determinada tarefa.

Vatagens do uso de funções:

1. Modularização
2. Reúso de código
3. Legibilidade

Você programa uma tarefa dentro de um bloco de código, dá um nome para essa função. Eventualmente pode ter uma lista de parâmetros e um valor de retorno dependendo do tipo de função.

Pode usar essa função quantas vezes quiser no seu programa ou até mesmo importar na forma de um módulo em outros programas.

Permite a modularização do programa. Você consegue tornar o seu código modular criando um bloco de códigos que realize uma determinada tarefa e usando posteriormente, promovendo o reúso de código. Não precisa ficar digitando o mesmo código diversas vezes.

A função é como se fosse uma caixa preta, você tem um valor de entrada e dentro da função executam-se instruções e  determinadas ações e resulta em um valor saída.

Ao se chamar uma função não precisamos saber exatamente como a função é implementada, como ela funciona, podendo, inclusive, ter sido feita por outra pessoa.

Existem dois tipos de funções em Python: (1) as funções internas, nativas e (2) as funções definidas pelo usuário, são aquelas que nós criamos ou outras pessoas.

Criamos uma função em Python usando a palavra reservada `def` de define.

```python
def <nome_função> ([argumentos]):
  <instruções>
```

```python
def mensagem():
  print('Bóson Treinamentos em Tecnologia')
  print('Curso completo de Python')

mensagem()
```

# Função com argumentos

Argumento é um valor que você passa para a função, como se fosse uma variável, um porteiro que recebe um valor e leva pra dentro da função.

```python
def soma(a, b):
  print(a + b)

soma(12, 7)
```

Quando precisamos armazenar o valor para ser usado posteriormente usamos a palavra reservada `return`.

```python
def mult(x, y):
  return(x * y)

a = 5
b = 8
c = mult(a, b)
print(f'O produto de {a} e {b} é {c}')
```

```python
def div(k, j):
  if j != 0:
    return(k / j)
  else:
    return 'Impossível dividir por zero!'

if __name__ == '__main__':
  a = int(input('Digite um número: '))
  b = int(input('Digite outro número: '))
  r = div(a, b)
  print(f'{a} dividido por {b} é igual a {r}')
```

# Programa para calcular quadrados de números

```python
def quadrado(val):
  quadrados = []
  for x in val:
    quadrados.append(x ** 2)
  return quadrados

if __name__ == '__main__':
  valores = [2, 5, 7, 9, 12]
  resultados = quadrado(valores)
  for g in resultados:
    print(g)
```
