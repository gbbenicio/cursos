# Operadores Lógicos AND, OR, NOT

Os operadores lógicos implementam a lógica booleana.

Trabalham com valores True e False.

Os principais operadores lógicos são AND, OR e NOT

## Tabela Verdade AND

| Condição A | Condição B | Resultado |
|------------|------------|-----------|
| False | False | False |
| True | False | False |
| False | True | False |
| True | True | True |


## Tabela Verdade OR

| Condição A | Condição B | Resultado |
|------------|------------|-----------|
| False | False | False |
| True | False | True |
| False | True | True |
| True | True | True |


## Tabela Verdade NOT

| Entrada | Saída |
|---------|-------|
| False | True |
| True | False |

NOT é um operador UNÁRIO, só usa um operando. Recebe apenas um valor na Entrada e inverte o valor na SAÍDA.


Consideremos uma situação hipotética: Em um evento esportivo, para participar, a pessoa tem que ter mais de 18 anos e tem altura superior a 1,70m. Como implementaríamos isso, essa verificação?

```python
idade = 25
altura = 1.75
```

```python
resultado = (idade >= 18) and (altura >= 1.70)
msg = 'Pode participar do evento? ' + str(resultado)

print(msg)
```

Programa de disparo de alarme em uma residência conectada a sensores em uma porta e janela.

f: fechado  
a: aberto

```python
porta = 'f'
janela = 'f'
```

```python
porta = 'a'
janela = 'f'
```

```python
porta = 'f'
janela = 'a'
```

```python
porta = 'a'
janela = 'a'
```

```python
alarme = (porta == 'a') or (janela == 'a')
msg = 'Alarme disparado? ' + str(alarme)
print(msg)
```

Programa para verificar se tem ou não dinheiro na carteira usando o operador lógico NOT.

```python
tem_dinheiro = False
```

```python
tem_dinheiro = not tem_dinheiro
```

```python
msg = 'Tem dinheiro? ' + str(tem_dinheiro)
print(msg)
```
