| Operador | Operação |
|----------|----------|
| + | Soma |
| - | Subtração |
| * | Multiplicação |
| / | Divisão real |
| // | Divisão inteira |
| % | Módulo (MOD) |
| ** | Potenciação |

| | Ordem de Precedência dos Operadores: |
|---|--|
|**1**|Parênteses|
|**2**|Potenciação|
|**3**|Multiplicação / Divisão|
|**4**|Soma / Subtração|
|**5**|Esquerda para a Direita|

```python
a = 10
b = 5
c = 11

print('a =', a)
print('b =', b)
print('c =', c)

# a + b
print('\nSoma:\n', 'a + b =', (a + b))

# a - b
print('\nSubtração:\n', 'a - b =', (a - b))

# b - a
print(' b - a =', (b - a))

# a * b
print('\nMultiplicação:\n', 'a * b =', (a * b))

# a / b
print('\nDivisão real:\n', 'a / b =', (a / b))

# b / a
print(' b / a =', (b / a))

# b // a
print('\nDivisão inteira:\n', 'b // a =', (b // a))

# a % b
print('\nMódulo:\n', 'a % b =', (a % b))

# c % b
print(' c % b =', (c % b))

# a % c
print(' a % c =', (a % c))

# a ** 2
print('\nPotenciação:\n', 'a ** 2 =', (a ** 2))

# a ** b
print(' a ** b =', (a ** b))

# a + b * c
print('\nCálculos:\n', 'a + b * c =', (a + b * c))

# (a + b) * c
print('\n(a + b) * c =', ((a + b) * c))
```

Podemos usar input para que o usuário informe os valores das variáveis.

Input sempre retorna uma string (cadeia de caracteres).

Então temos que usar operadores de Casting para converter string no tipo desejado. No caso usamos o casting int.

```python
x = int(input('Digite um número: '))
y = int(input('Digite outro número: '))

z = x + y

print('A soma dos dois números é', z)
```
